```python
# 1.1

import pandas as pd
import json

with open("definitivo.json", "r") as f:
    musica_json = json.load(f)

artistas = []

for artista in musica_json:
    genero = artista["genre"]
    año = artista["year"]
    
    if genero in ["rock", "flamenco", "indie", "latin", "rap"] and 2020 <= año <= 2025:
        
        artistas.append({
            "id": artista["id"],
            "artist_name": artista["artist_name"],
            "track_name": artista["track_name"],
            "genre": genero,
            "year": año
        })

data_music = pd.DataFrame(artistas)
data_music.head()

data_music.to_csv("data_music_filtrado.csv", index=False, encoding ="utf-8-sig")
```
